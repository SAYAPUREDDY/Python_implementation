# Getting started

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# plotting without line
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
plt.plot(xpoints, ypoints,'o')
plt.show()
```

```python
# multiple points
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# if we give only one set of points it will take automatically one set of values as 0,1,2,3
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints)
plt.show()
```

# matplotlib markers

```python
# we can mark each point with a circle
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
# marking each point with star
# can use lot of other symbols check the tutorials
# https://www.w3schools.com/python/matplotlib_markers.asp
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'X')
plt.show()
```

# Format strings fmt

```python
# : for dotted line
# - solid line
# -. dashed/dotted line
# color reference : r-red,g-green,b-blue,c-cyan,m-magenta,y-yellow,k-black,w-white
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o--k')
plt.show()
```

# Marker size 

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 15)
plt.show()
```

# marker edge color

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'k')
plt.show()
```

# marker face color

```python
# we can use hexadecimal color values in place of strings too
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'c',mec='k')
plt.show()
```

# line style

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,linestyle='dotted')
plt.show()
```

```python
# can use shorter syntax too
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,ls='dashed')
plt.show()
```

```python
# shorter syntax
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,ls=':')
plt.show()
```

```python
# line color
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, color = 'r')
plt.show()
```

```python
# shorter syntax
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, c = 'r')
plt.show()
```

```python
# use hexadecimal values too
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, color = '#4CAF50')
plt.show()
```

# Line width

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linewidth = '10')
plt.show()
```

# set font properties for Title and Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.show()
```

# Display multiple plots

```python
# plot:1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1,2,1)
plt.plot(x,y)

# plot:2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```

```python
# drawing two plots on top of each other

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
# adding the title 
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

```python
# adding the super title for both of the graphs
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle('MY SHOP')
plt.show()
```

# colormap

```python
# we will use the keyword cmap and can use inbuilt viridis cmap available
# define the array of values for color and use the cmap\
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.show()
```

```python
# there are lot of color maps available
# check the https://www.w3schools.com/python/matplotlib_scatter.asp
# including the colorbar in the plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

# Aplha

```python
# we can adjust the trasperancy of points
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

# combining the color size and alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Pie charts

```python
# a simple pie chart
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

# labels

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
# start angle
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,startangle=90)
plt.show()
```

```python
# explode
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# can add a shadow too
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

# colors

```python
# we can set colors to each wedges
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
# can use legend
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
# legend with header
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
